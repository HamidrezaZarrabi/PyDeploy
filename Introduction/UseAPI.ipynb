# Face Generation


```python
import requests
import numpy as np
import matplotlib.pyplot as plt
import json
import cv2
```

```python
url = "https://thispersondoesnotexist.com"
response = requests.get(url=url)
response.status_code
```

```python
file_data = np.frombuffer(response.content, dtype='uint8')
img = cv2.imdecode(file_data, cv2.IMREAD_UNCHANGED)
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
plt.axis(False)
plt.show()
```

# Plant Identification

```python
!pip install -q python-dotenv
```

```python
import requests
import dotenv
import os
```

```python
url = "https://my-api.plantnet.org/v2/identify/all"
env = dotenv.load_dotenv()
APIKey = os.getenv("APIKey_Plant")
ImgPath_1 = "DutchRose_1.jpeg"
ImgPath_2 = "DutchRose_2.jpeg"
Img_1 = open(ImgPath_1, 'rb')
Img_2 = open(ImgPath_2, 'rb')
headers = {}
params = {"api-key": APIKey}
Files = [
    ('images', (ImgPath_1, Img_1)),
    ('images', (ImgPath_2, Img_2))
]
```

```python
response = requests.post(url=url, headers=headers, params=params, files=Files)
response.status_code
```

```python
response.json()
```
